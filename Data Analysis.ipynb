```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('P RATE.csv')
ftr_data = pd.read_csv('fertility.csv')
elderly_data = pd.read_csv('Elderly pp.csv')
young_data = pd.read_csv('Young pp.csv')
```

```python
data
# data: Column added with Countryu name, and combined together.
```

```python
# compare the population growth in each country
```

```python
korea = data[data.Country == 'SK']
#saving each country data into name
```

```python
japan = data[data.Country == 'Japan']
#saving each country data into name
```

```python
china = data[data.Country == 'China']
#saving each country data into name
```

```python
plt.plot(korea.Year, korea.GrowthRate)
plt.plot(japan.Year, japan.GrowthRate)
plt.plot(china.Year, china.GrowthRate)
plt.legend(['Korea', 'Japan', 'China'])
plt.xlabel('year')
plt.ylabel('Growth Rate')
plt.show()
# This shows years historical comparison of population groth rate between three countries around Korea.
# As a result all of three country show its dropping growth rate and South Korea has the highest slope
# of dropping since 1960
#South Korea Had Korean war during 1950 and had huge spike on 1960s due to economic/financial recovery
```

```python
plt.plot(korea.Year, korea.TotalPopulation)
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
#We can also see from this graph that increasing slope has been decreased over time.
#We will not be comparing all population since number gaps are big.
#As we can see no matter what age groups people are in, it has increasing slope line.
```

```python
korea
# Korea's Specific data from csv
```

```python
#Growth rate untill now since 1955, results in 5years cycle
korea.GrowthRate
```

```python
df = korea.GrowthRate
sum1 = df.sum()
print(sum1)
avg_korea_growth = sum1/18
print(avg_korea_growth)
cgr = df.diff() 
print(cgr)
# difference in each year's rate, Growth Rate is in positive numbers
# However the difference in each year show how numbers are declining
```

```python
plt.scatter(korea.Year,korea.GrowthRate)
plt.xlabel('Year')
plt.ylabel('Growth rate')
x = data['Year']
y = data['GrowthRate']
z = np.polyfit(x,y,1)
p = np.poly1d(z)
plt.plot(x,p(x), "r--")
plt.show()

# as we can see below graph total slope of the line shows its in declining state.
# we had a spike jump of growth rate in 1960 due to economic development surge after korean war in 1950
# Average trendline indicates declining as well.
```
